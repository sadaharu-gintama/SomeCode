```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
def f(x):
    return np.sin(x) + 0.5 * x
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
```

```python
plt.plot(x, f(x), 'b')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

# Regression #

```python
reg = np.polyfit(x, f(x), deg = 1)
ry = np.polyval(reg, x)
```

```python
plt.plot(x, f(x), 'b', label = 'f(x)')
plt.plot(x, ry, 'r.', label = 'regression')
plt.legend(loc = 0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

## Fit with 7th degree ##

## lstsq##

```python
matrix = np.zeros((4, len(x)))
matrix[3, :] = x ** 3
matrix[2, :] = x ** 2
matrix[1, :] = x
matrix[0, :] = 1
```

```python
reg = np.linalg.lstsq(matrix.T, f(x))[0]
```

```python
reg
```

```python
ry = np.dot(reg, matrix)
```

```python
plt.plot(x, f(x), 'b', label = 'f(x)')
plt.plot(x, ry, 'r.', label = 'regression')
plt.legend(loc = 0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
matrix[3, :] = np.sin(x)
reg = np.linalg.lstsq(matrix.T, f(x))[0]
ry = np.dot(reg, matrix)
```

```python
plt.plot(x, f(x), 'b', label = 'f(x)')
plt.plot(x, ry, 'r.', label = 'regression')
plt.legend(loc = 0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
np.allclose(f(x), ry)
```

## Multi-dimension ##

```python
def fm((x, y)):
    return np.sin(x) + 0.25 * x + np.sqrt(y) + 0.05 * y ** 2
```

```python
x = np.linspace(0, 10, 20)
y = np.linspace(0, 10, 20)
X, Y = np.meshgrid(x, y)
Z = fm((X, Y))
x = X.flatten()
y = Y.flatten()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl

fig = plt.figure(figsize = (9,6))
ax = fig.gca(projection = '3d')
surf = ax.plot_surface(X, Y, Z, rstride = 2, cstride = 2,
                      cmap = mpl.cm.coolwarm, linewidth = .5, antialiased = True)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
fig.colorbar(surf, shrink = .5, aspect = 5)
```

```python
matrix = np.zeros((len(x), 7))
matrix[:, 6] = np.sqrt(y)
matrix[:, 5] = np.sin(x)
matrix[:, 4] = y ** 2
matrix[:, 3] = x ** 2
matrix[:, 2] = y
matrix[:, 1] = x
matrix[:, 0] = 1
```

```python
import statsmodels.api as sm
model = sm.OLS(fm((x, y)), matrix).fit()
```

```python
model.rsquared
```

```python
a = model.params
```

```python
a
```

```python
def reg_func(a, (x, y)):
    f6 = a[6] * np.sqrt(y)
    f5 = a[5] * np.sin(x)
    f4 = a[4] * y ** 2
    f3 = a[3] * x ** 2
    f2 = a[2] * y
    f1 = a[1] * x
    f0 = a[0]
    return (f0 + f1 + f2 + f3 + f4 + f5 + f6)
```

```python
RZ = reg_func(a, (X, Y))
```

```python
fig = plt.figure(figsize = (9,6))
ax = fig.gca(projection = '3d')
surf1 = ax.plot_surface(X, Y, Z, rstride = 2, cstride = 2,
                       cmap = mpl.cm.coolwarm, linewidth = .5, antialiased = True)
surf2 = ax.plot_wireframe(X, Y, RZ, rstride = 2, cstride = 2, label = 'regression')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
ax.legend()
fig.colorbar(surf, shrink = .5, aspect = 5)
```

# Convex Optimization#

```python
def fm((x, y)):
    return (np.sin(x) + 0.05 * x ** 2 + np.sin(y) + 0.05 * y ** 2)
```

```python
x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x, y)
Z = fm((X, Y))
```

```python
fig = plt.figure(figsize = (9,6))
ax = fig.gca(projection = '3d')
surf = ax.plot_surface(X, Y, Z, rstride = 2, cstride = 2, 
                       cmap = mpl.cm.coolwarm, linewidth = .5, antialiased = True)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
fig.colorbar(surf, shrink = .5, aspect = 5)
```

```python
def fo((x, y)):
    z = np.sin(x) + 0.05 * x ** 2 + np.sin(y) + 0.05 * y ** 2
    if output == True:
        print '%8.4f %8.4f %8.4f' % (x, y, z)
    return z
```

## Optimization with constraints ##

```python
from math import sqrt
def Eu((s, b)):
    return -(.5 * sqrt(s * 15 + b * 5) + .5 * sqrt(s * 5 + b * 12))
cons = ({'type': 'ineq', 'fun' : lambda (s, b): 100 - s * 10 - b * 10})
bnds = ((0, 1000), (0, 1000))
```

```python
result = spo.minimize(Eu, [5, 5], method = 'SLSQP', bounds = bnds, constraints = cons)
```

```python
result
```

```python
np.dot(result['x'], [10, 10])
```
