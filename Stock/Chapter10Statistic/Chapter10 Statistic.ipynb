```python
x0 = 0.05
kappa = 3.
theta = .02
sigma = .1
```

```python
def srd_euler():
    xh = np.zeros((M + 1, I))
    x1 = np.zeros_like(xh)
    xh[0] = x0
    x1[0] = x0
    for t in range(1, M + 1):
        xh[t] = (xh[t - 1] + kappa * (theta - np.maximum(xh[t - 1], 0)) * dt
                + sigma * np.sqrt(np.maximum(xh[t - 1], 0)) * np.sqrt(dt) * npr.standard_normal(I))
    x1 = np.maximum(xh, 0)
    return x1
x1 = srd_euler()
```

```python
plt.hist(x1[-1], bins = 50)
plt.xlabel('value')
plt.ylabel('frequency')
plt.grid(True)
```

```python
plt.plot(x1[:, :10], lw = 1.5)
plt.xlabel('Time')
plt.ylabel('index level')
plt.grid(True)
```

```python
def srd_exact():
    x2 = np.zeros((M + 1, I))
    x2[0] = x0
    for t in range(1, M + 1):
        df = 4 * theta * kappa / sigma ** 2
        c = (sigma ** 2 * (1 - np.exp(-kappa * dt))) / (4 * kappa)
        nc = np.exp(-kappa * dt) / c * x2[t - 1]
        x2[t] = c * npr.noncentral_chisquare(df, nc, size = I)
    return x2
x2 = srd_exact()
```

```python
def gen_sn(M, I, anti_paths = True, mo_match = True):
    if anti_paths:
        sn = npr.standard_normal((M + 1, I / 2))
        sn = np.concatenate((sn, -sn), axis = 1)
    else:
        sn = npr.standard_normal((M + 1, I))
    if mo_match:
        sn = (sn - sn.mean()) / sn.std()
    return sn
```

# Estimation#

## European Option##

```python
S0 = 100.
r = .05
sigma = .25
T = 1.0
I = 50000

def gbm_mcs_stat(K):
    sn = gen_sn(1, I)
    ST = S0 * np.exp((r - .5 * sigma ** 2) * T + sigma * np.sqrt(T) * sn[1])
    hT = np.maximum(ST - K, 0)
    C0 = np.exp(-r * T) * 1 / I * np.sum(hT)
    return C0
```

```python
gbm_mcs_stat(105.)
```

```python
M = 50
def gbm_mcs_dyna(K, option = 'call'):
    dt = T / M
    S = np.zeros((M + 1, I))
    S[0] = S0
    sn = gen_sn(M, I)
    for t in range(1, M + 1):
        S[t] = S[t - 1] * np.exp((r - .5 * sigma ** 2) * dt + sigma * np.sqrt(dt) * sn[t])
    if option == 'call':
        hT = np.maximum(S[-1] - K, 0)
    else:
        hT = np.maximum(K - S[-1], 0)
    C0 = np.exp(-r * T) * 1 / I * np.sum(hT)
    return C0
```

## American Option##

```python
def gbm_mcs_amer(K, option = 'call'):
    dt = T / M
    df = np.exp(-r * dt)
    S = np.zeros((M + 1, I))
    S[0] = S0
    sn = gen_sn(M, I)
    for t in range(1, M + 1):
        S[t] = S[t - 1] * np.exp((r - .5 * sigma ** 2) * dt + sigma * np.sqrt(dt) * sn[t])
    if option == 'call':
        h = np.maximum(S - K, 0)
    else:
        h = np.maximum(K - S, 0)
    
    V = np.copy(h)
    for t in range(M - 1, 0, -1):
        reg = np.polyfit(S[t], V[t + 1] * df, 7)
        C = np.polyval(reg, S[t])
        V[t] = np.where(C > h[t], V[t + 1] * df, h[t])
    C0 = df * 1 / I * np.sum(V[1])
    return C0
```
