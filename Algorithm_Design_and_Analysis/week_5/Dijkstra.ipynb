```python
import numpy as np
```

```python
def loadData(fileName):
    branch = list()
    cost = list()
    with open(fileName) as fptr:
        for line in fptr.readlines():
            ll = line.strip().split()[1:]
            branch.append([int(l.split(",")[0]) - 1 for l in ll])
            cost.append([float(l.split(",")[1]) for l in ll])
    return branch, cost
```

```python
branch, cost = loadData("dijkstraData.txt")
```

```python
cost
```

```python
def dijkstra(st, br, weight):
    shortest_path = [0] * 200
    x = [st]
    for i in range(1, 200):
        shortest = np.inf
        point = -1
        for p in x:
            end = [(b,c) for b, c in zip(br[p], weight[p]) if b not in x]
            for b, c in end:
                current = shortest_path[p] + c
                if current < shortest:
                    shortest = current
                    point = b
        shortest_path[point] = shortest
        x.append(point)
    return shortest_path
```

```python
st = dijkstra(0, branch, cost)
```
