```python
time_zones = [r['tz'] for r in records]
```

```python
time_zones = [r['tz'] for r in records if 'tz' in r]
```

```python
time_zones[:10]
```

```python
def count_tz(time_zones):
    count = dict()
    for t in time_zones:
        if t in count:
            count[t] += 1
        else:
            count[t] = 1
    return count
```

```python
count = count_tz(time_zones)
```

```python
count[0]
```

# 1880-2010 Baby names

```python
names1880 = pd.read_csv('names/yob1880.txt', names = ['name', 'sex', 'births'])
```

```python
names1880[:5]
```

```python
names1880.groupby('sex').births.sum()
```

```python
years = range(1880, 2010)
pieces = list()
columns = ['name', 'sex', 'births']
for year in years:
    path = 'names/yob%d.txt' % year
    frame = pd.read_csv(path, names = columns)
    frame['year'] = year
    pieces.append(frame)
```

```python
names = pd.concat(pieces, ignore_index=True)
```

```python
names[:5]
```

```python
total_births = names.pivot_table('births', index = 'year', columns = 'sex', aggfunc=sum)
```

```python
total_births
```

```python
total_births.plot(title = 'total births by sex and year')
```

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group

names = names.groupby(['year', 'sex']).apply(add_prop)
```

```python
names.tail()
```

```python
np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)
```

```python
def get_top1000(group):
    return group.sort_index(by = 'births', ascending = False)[:1000]

grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
```

```python
top1000.tail()
```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
```

```python
total_births = top1000.pivot_table('births', index = 'year', columns = 'name', aggfunc = sum)
```

```python
total_births
```

```python
subset = total_births[['John', 'Harry', 'Mary', 'Marilyn']]
```

```python
subset.plot(subplots = True, figsize = (12, 10), grid = True, title = 'Number of births per year')
```

```python
table.plot(title = 'Sum of table1000.prop by year and sex', yticks = np.linspace(0, 1.2, 13), xticks = range(1880, 2020, 10))
```

```python
df = boys[boys.year == 2009]
```

```python
prop_cumsum = df.sort_index(by = 'prop', ascending = False).prop.cumsum()
```

```python
def get_quantile_count(group, q = 0.5):
    group = group.sort_index(by='prop', ascending = False)
    return group.prop.cumsum().searchsorted(q)[0] + 1

diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
```

```python
diversity.head()
```

```python
diversity.plot(title = 'Number of popular names in top 50%')
```
