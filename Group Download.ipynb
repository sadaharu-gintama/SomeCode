```python
from bs4 import BeautifulSoup
import urllib2
```

```python
url = urllib2.urlopen('http://ocw.mit.edu/courses/engineering-systems-division/esd-934-engineering-economics-and-regulation-of-the-electric-power-sector-spring-2010/lecture-notes/')
```

```python
soup = BeautifulSoup(url.read(), 'lxml')
```

```python
print(soup.prettify())
```

```python
listOfPDF = list()
for link in soup.find_all('a'):
    urls = link.get('href')
    if urls[-4:] == '.pdf':
        listOfPDF.append(urls)
```

```python
print listOfPDF
```

```python
for pdf in listOfPDF:
    fileName = pdf.split('/')[-1]
    with file(fileName, 'wb') as f:
        f.write(urllib2.urlopen('http://ocw.mit.edu' + pdf).read())
```
